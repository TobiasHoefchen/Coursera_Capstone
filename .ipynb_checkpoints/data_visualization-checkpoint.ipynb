```python
processed_data_dict = pd.read_pickle("./data/clustered_data_dict.p")
processed_data_df = pd.read_pickle("./data/clustered_data_df.txt")
```

## Creating a map

```python
# create map of New York using latitude and longitude values
map_fig = folium.Map(location=[43.6534817, -79.3839347], zoom_start=10)

# set color scheme for the clusters
kclusters = len(processed_data_df["Cluster Labels"].unique())
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(processed_data_df["Neighborhood Latitude"], processed_data_df["Neighborhood Longitude"], processed_data_df["Neighborhood"], processed_data_df['Cluster Labels']):
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=f"<i>{str(poi)}, <b> Cluster {str(cluster)}</b></i>",
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_fig)
    
map_fig        
```

## Inspect Clusters

```python
def get_piedata(data, catColDict):
    data = data.iloc[:, 3:]
    
    #get values
    values = data.loc[-1]
    
    #get labels
    labels = data.columns
    n_max = sum(values)
    labels_with_perc = [label + f", {np.round(data[label].loc[-1]/n_max*100, 2)}%" for label in data.columns]
    
    #get color
    label_colors = [catColDict[category] for category in labels]
    
    return values, labels_with_perc, label_colors
```

```python
def visualize_cluster(data, cluster, catColDict):
    values, labels, label_colors = get_piedata(processed_data_dict[cluster], catColDict)
    
    fig, ax = plt.subplots(figsize =(10, 7))
    plt.pie(values, labels = labels, colors = label_colors)
    ax.set_title(f"Percentage of venues in Neighborhoods of {cluster}", size=17)
    plt.show()
    return
```

```python
#create colors dictionary
categories = list(processed_data_df.columns[4:])
colors_array = cm.rainbow(np.linspace(0, 1, len(categories)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
categories_color_dict = {category: color for (category, color) in zip(categories, rainbow)}
```

```python
for clusterNr in range(processed_data_df["Cluster Labels"].min(), processed_data_df["Cluster Labels"].max()):
    visualize_cluster(processed_data_dict, f"Cluster {clusterNr}", categories_color_dict)
```
